```python
import pandas as pd
import numpy as np
SEED = 20

np.random.seed(SEED)

train_uri = 'https://raw.githubusercontent.com/adriano06-tech/kaggle_titanic_ML/main/data/train.csv'
train = pd.read_csv(train_uri)
train.head()
```

```python
train.drop(columns = ['PassengerId', 'Name'], inplace = True)
train.head()
```

```python
train['Sex'] = train['Sex'].map({'male': 0, 'female': 1})
train.head()
```

```python
train['Cabin'] = train['Cabin'].fillna('NULL')
train.head()
```

```python
train['Age'] = train['Age'].fillna(train['Age'].median())
train.head()
```

```python
train_X = train[train.columns[1:]]
train_Y = np.array(train['Survived'])

print(f'X: {train_X.shape} ||| Y: {train_Y.shape}')
```

```python
ticket_remap = {}
ticket_i = 0
cabin_remap = {}
cabin_i = 0
embarked_remap = {}
embarked_i = 0

for index, row in train_X.iterrows():
    if row['Ticket'] not in ticket_remap:
        ticket_remap[row['Ticket']] = ticket_i
        ticket_i += 1
        
    if row['Cabin'] not in cabin_remap:
        cabin_remap[row['Cabin']] = cabin_i
        cabin_i += 1
    
    if row['Embarked'] not in embarked_remap:
        embarked_remap[row['Embarked']] = embarked_i
        embarked_i += 1
        

train_X['Ticket'] = train_X['Ticket'].map(ticket_remap)
train_X['Cabin'] = train_X['Cabin'].map(cabin_remap)
train_X['Embarked'] = train_X['Embarked'].map(embarked_remap)

train_X    
```

```python
from sklearn.preprocessing import StandardScaler

train_X_columns_labels = train_X.columns

scaler = StandardScaler()

train_X = scaler.fit_transform(train_X)

train_X = pd.DataFrame(train_X)
train_X.columns = train_X_columns_labels
train_X.head()
```

```python
test_uri = 'https://raw.githubusercontent.com/adriano06-tech/kaggle_titanic_ML/main/data/test.csv'
test = pd.read_csv(test_uri)
test.head()
```

```python
gender_submission_uri = 'https://raw.githubusercontent.com/adriano06-tech/kaggle_titanic_ML/main/data/gender_submission.csv'
gender_submission = pd.read_csv(gender_submission_uri)
gender_submission.head()
```

```python
test_Y = []

for index, row in test.iterrows():
    select = gender_submission['PassengerId'] == row['PassengerId']
    survived = gender_submission[select].values[0][1]
    test_Y.append(survived)
    
test_Y = np.array(test_Y)
test_Y
```

```python
test.drop(columns = ['PassengerId', 'Name'], inplace = True)
test.head()
```

```python
test['Sex'] = test['Sex'].map({'male': 0, 'female': 1})
test.head()
```

```python
test['Cabin'] = test['Cabin'].fillna('NULL')
test.head()
```

```python
test['Age'] = test['Age'].fillna(test['Age'].median())
test.head()
```

```python
test_X = test[test.columns[:]]

print(f'X: {test_X.shape} ||| Y: {test_Y.shape}')
```

```python
ticket_remap = {}
ticket_i = 0
cabin_remap = {}
cabin_i = 0
embarked_remap = {}
embarked_i = 0

for index, row in test_X.iterrows():
    if row['Ticket'] not in ticket_remap:
        ticket_remap[row['Ticket']] = ticket_i
        ticket_i += 1
        
    if row['Cabin'] not in cabin_remap:
        cabin_remap[row['Cabin']] = cabin_i
        cabin_i += 1
    
    if row['Embarked'] not in embarked_remap:
        embarked_remap[row['Embarked']] = embarked_i
        embarked_i += 1
        

test_X['Ticket'] = test_X['Ticket'].map(ticket_remap)
test_X['Cabin'] = test_X['Cabin'].map(cabin_remap)
test_X['Embarked'] = test_X['Embarked'].map(embarked_remap)

test_X    
```

```python
from sklearn.preprocessing import StandardScaler

test_X_columns_labels = test_X.columns

scaler = StandardScaler()

test_X = scaler.fit_transform(test_X)

test_X = pd.DataFrame(test_X)

test_X.columns = test_X_columns_labels
test_X.head()
```

```python
test_X = test_X.dropna()
test_X = np.array(test_X)
```

```python
def calculate_accuracy(classifier):
    predict_labels = classifier.predict(test_X)
    
    
    accuracy_score = 0

    for i in range(len(predict_labels)):
        if(predict_labels[i] == test_Y[i]):
            accuracy_score += 1

    accuracy = accuracy_score / len(predict_labels) * 100
    
    return accuracy
    
```

```python
from sklearn.svm import LinearSVC

classifier_linear_svc = LinearSVC()

classifier_linear_svc.fit(train_X, train_Y)

linear_svc_accuracy = calculate_accuracy(classifier_linear_svc)
print(f'The LinearSVC algorithm score was {round(linear_svc_accuracy, 2)}%')
```

```python
from sklearn.ensemble import RandomForestClassifier

classifier_random_forest = RandomForestClassifier()

classifier_random_forest.fit(train_X, train_Y)

random_forest_accuracy = calculate_accuracy(classifier_random_forest)
print(f'The Random Forest Accuracy algorithm score was {round(random_forest_accuracy, 2)}%')
```

```python
from sklearn.svm import SVC

classifier_SVC = SVC()

classifier_SVC.fit(train_X, train_Y)

svc_accuracy = calculate_accuracy(classifier_SVC)
print(f'The SVC algorithm score was {round(svc_accuracy, 2)}%')
```
